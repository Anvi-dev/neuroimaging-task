```python
#Task 2: DICOM Processing


import os
import pydicom
import numpy as np

dicom_folder = "/mnt/c/Users/sadas/Downloads/dicom/"

dicoms = []

for file in os.listdir(dicom_folder):
    path = os.path.join(dicom_folder, file)
    try:
        dcm = pydicom.dcmread(path, force=True)
        dicoms.append(dcm)
    except:
        continue

print("Total DICOM slices:", len(dicoms))

# Extract metadata
dcm0 = dicoms[0]
print("Patient ID:", dcm0.PatientID)
print("Study Date:", dcm0.StudyDate)
print("Modality:", dcm0.Modality)

# Sort slices
dicoms_sorted = sorted(dicoms, key=lambda x: int(x.InstanceNumber))

print("First Instance:", dicoms_sorted[0].InstanceNumber)
print("Last Instance:", dicoms_sorted[-1].InstanceNumber)
```
